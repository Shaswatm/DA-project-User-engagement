# Plot and Inferences

```python
%matplotlib inline
#%config InlineBackend.figure_format='retina'

from __future__ import absolute_import, division, print_function
import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib.pyplot import GridSpec
import seaborn as sns
import numpy as np
import pandas as pd
import datetime as dt

sns.set_context("poster", font_scale=1.3)
```

```python
#import data csv to keep time format specify else imported as int
user_df = pd.read_csv("clean_df.csv" , parse_dates= ['time_stamp'])
```

```python
#check the imported dataframe
user_df.head(2)
```

```python
user_df = user_df.iloc[:,1:]# removing the 1st colummn
```

## VISUALIZE PLOTS

### Users Count by Mobile brand

```python
MIDSIZE = (18,22)
fig, ax = plt.subplots(figsize=MIDSIZE)
user_df.mobile_brand_name.value_counts(dropna=False, ascending=True).plot(kind='barh', ax=ax)
ax.set_ylabel("mobile_brand_name")
ax.set_xlabel("Counts")
```

### Uses by Week days 

```python
MIDSIZE = (12, 8)
fig, ax = plt.subplots(figsize=MIDSIZE)
user_df.time_stamp.dt.weekday_name.value_counts(dropna=False, ascending=True).plot(kind='barh', ax=ax)
ax.set_ylabel("week_days")
ax.set_xlabel("Counts")
fig.tight_layout()
```

### User by weekday trend

```python
SMALLSIZE = (12,5)
fig, ax = plt.subplots(figsize=SMALLSIZE)
user_df.time_stamp.dt.weekday_name.value_counts().plot()# this is sorted by values 
```

### User by hours trend¶

```python
SMALLSIZE = (12,5)
fig, ax = plt.subplots(figsize=SMALLSIZE)
user_df.time_stamp.dt.hour.value_counts().sort_index().plot()
```
