## Visulalize features of univariate features

### Users Count by Mobile brand

```python
MIDSIZE = (18,22)
fig, ax = plt.subplots(figsize=MIDSIZE)
user_df.mobile_brand_name.value_counts(dropna=False, ascending=True).plot(kind='barh', ax=ax)
ax.set_ylabel("mobile_brand_name")
ax.set_xlabel("Counts")
#fig.tight_layout()
```

### Uses by Week days 

```python
MIDSIZE = (12, 8)
fig, ax = plt.subplots(figsize=MIDSIZE)
user_df['week_days'].value_counts(dropna=False, ascending=True).plot(kind='barh', ax=ax)
ax.set_ylabel("week_days")
ax.set_xlabel("Counts")
fig.tight_layout()
```

```python
#ERROR Debugging: dur to datetime imported (csv to dataframe) in int64 convert to datetime object

import datetime as dt
user_df.time_stamp = pd.to_datetime(user_df.time_stamp)
```

```python
# recheck and chages
user_df.dtypes
```

### User by weekday trend

```python
SMALLSIZE = (12,5)
fig, ax = plt.subplots(figsize=SMALLSIZE)
user_df.time_stamp.dt.weekday_name.value_counts().plot()# this is sorted by values 
#user_df.time_stamp.dt.day.plot?
```

### User by hours trend

```python
SMALLSIZE = (12,5)
fig, ax = plt.subplots(figsize=SMALLSIZE)
user_df.time_stamp.dt.hour.value_counts().sort_index().plot()
```
